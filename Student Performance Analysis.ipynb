#### **KNN Imputation**


**K-Nearest Neighbors (KNN) imputation fills missing values by finding the k most similar rows based on other features. It replaces missing values with the average (for numerical data) or most common category (for categorical data) among the neighbors.Pros: Maintains data structure, useful for small missing data.Cons: Slow for large datasets, sensitive to k choice.**


```python
#Fill nulls using KNN
imputed_columns = ['Attendance (%)', 'Assignments_Avg']
knn_imputer = KNNImputer(n_neighbors=5)

df[imputed_columns] = knn_imputer.fit_transform(df[imputed_columns])
```

```python
#Continuous features explorarion function
def continuous_explore(column):
    column = df[column]
    length = len(column)
    minimum = column.min()
    maximum = column.max()
    data_range = maximum - minimum
    mean = column.mean()
    median = column.median()
    std = column.std()
    q1 = column.quantile(0.25)
    q3 = column.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    outliers_count = ((column < lower) | (column > upper)).sum()
    outliers_per = (outliers_count / len(column)) * 100

    #Create dataframe for results
    continuous_result = pd.DataFrame({
        'Metric' : ['Name', 'Length', 'Min value', 'Max value', 'Range', 'Mean', 'Median', 'Standard deviation', 'Q1', 'Q3', 
                    'IQR', 'Lower band', 'Upper band', 'Outliers count', 'Outliers percentage'],

        'Value' : [column.name, length, minimum, maximum, data_range, mean, median, std, q1, q3, 
                   iqr, lower, upper, outliers_count, outliers_per]
    })

    continuous_result = (continuous_result.reset_index(drop=True)).style.set_properties(**{
        'text-align': 'center'}).set_table_styles([{'selector': 'th', 'props': [('text-align', 'center')]}])

    return continuous_result
```

#### **Histogram**

**A histogram is a graphical representation of the distribution of a dataset. It groups data into bins (intervals) and displays the frequency of values within each bin. The x-axis represents the data range, while the y-axis represents the count or frequency. Histograms help visualize the shape of the data distribution, detect skewness, and identify outliers.**


#### **Box Plot**

**A box plot (or box-and-whisker plot) is a statistical visualization that summarizes the distribution of a dataset using five key values: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. It helps identify outliers, data spread, and skewness. The box represents the interquartile range (IQR), while whiskers extend to the minimum and maximum values within 1.5 * IQR. Outliers appear as individual points outside the whiskers.**

```python
#Continuous features visualization function
def continuous_plot(column):
    fig, ax = plt.subplots(1, 2, figsize=(15, 5))
    color = random.choice(blues)
    
    #Histogram
    sns.histplot(x=column, data=df, ax=ax[0], color=color, kde=True, bins=20)
    ax[0].set_title(f'{column} Histogram', fontsize=18)   
    ax[0].tick_params(axis='both', which='major', labelsize=15)
    ax[0].set_xlabel(column, fontsize=15)
    ax[0].set_ylabel('Count', fontsize=15) 
    
    #Box plot
    sns.boxplot(x=column, data=df, ax=ax[1], color=color)
    ax[1].set_title(f'{column} Box Plot', fontsize=18)
    ax[1].tick_params(axis='both', which='major', labelsize=15)
    ax[1].set_xlabel(column, fontsize=15)

    plt.tight_layout()  
    plt.show()
```

```python
#Categorical features explorarion function
def categorical_explore(column):
    column = df[column]  
    length = len(column)  
    mode = column.mode()[0] if not column.mode().empty else None  
    category_number = column.nunique()  
    category_counts = column.value_counts()  
    categories = category_counts.index.tolist()  
    category_count_values = category_counts.values.tolist()
    
    #Create dataframe for results
    categorical_result = pd.DataFrame({
        'Metric': ['Name', 'Length', 'Mode', 'Number of categories', 'Categories', 'Categoris count'],
        
        'Value': [column.name, length, mode, category_number, categories, category_count_values ]
    })

    categorical_result = (categorical_result.reset_index(drop=True)).style.set_properties(**{
        'text-align': 'center'}).set_table_styles([{'selector': 'th', 'props': [('text-align', 'center')]}])
    
    return categorical_result
```

#### **Bar Plot**

**A bar plot is a graphical representation used to compare categorical data. It displays categories on the x-axis and their corresponding values (counts or aggregated metrics) on the y-axis using rectangular bars. The height of each bar represents the frequency or magnitude of the data. Bar plots can be grouped or stacked when comparing multiple categories within a dataset. They are useful for visualizing distributions, trends, and comparisons in categorical variables.**

```python
#Categorical features visualization function
def categorical_plot(column):
    plt.figure(figsize=(10, 6))
    
    df_counts = df[column].value_counts().reset_index()
    df_counts.columns = [column, 'count']
    df_counts['percentage'] = (df_counts['count'] / df_counts['count'].sum()) * 100
    
    ax = sns.barplot(x=column, y='count', data=df_counts, palette=reds, edgecolor='black', order=df_counts[column])  
    plt.xlabel(column, fontsize=15)
    plt.ylabel('Count', fontsize=15)
    plt.title(f'{column} Distribution', fontsize=18)

    #To add percentage on bars
    for p, percentage in zip(ax.patches, df_counts['percentage']):
        height = p.get_height()
        ax.text(p.get_x() + p.get_width() / 2, height + 0.5, f'{percentage:.1f}%', 
                ha='center', va='bottom', fontsize=11, color='black', fontweight='bold')

    plt.show()
```

#### **Scatter Plot**

**A scatter plot is a type of visualization used to show the relationship between two numerical variables. Each point on the plot represents an observation, with its position determined by the values of the two variables. Scatter plots help identify patterns, trends, correlations, and potential outliers in the data. If the points form a clear pattern (e.g., upward or downward trend), it suggests a relationship between the variables. They are particularly useful in regression analysis and exploratory data analysis (EDA).**

```python
#Scatter plot function to visualize 2 continuous features and 1 categorical feature
def scatter_plot(x_col, y_col, hue_col):
    
    plt.figure(figsize=(8, 6))
    palette = sns.color_palette('coolwarm', n_colors=df[hue_col].nunique())

    scatter = sns.scatterplot(
        data=df, x=x_col, y=y_col, hue=hue_col, palette=palette, edgecolor='black'
    )

    plt.xlabel(x_col, fontsize=12)
    plt.ylabel(y_col, fontsize=12)
    plt.title(f"{y_col} vs {x_col} vs {hue_col})", fontsize=14)
    
    plt.legend(title=hue_col, bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()
```

```python
#Bar plot function to visualize 2 categorical columns and continous feature
def bar_plot(x_col, y_col, hue_col):
    
    plt.figure(figsize=(8, 6))
    unique_values = df[hue_col].unique()
    
    color_list = ['#03396C', '#F27B92', '#B3CDE0', '#CC0033', '#6497B1']
    custom_palette = {val: color_list[i % len(color_list)] for i, val in enumerate(unique_values)}
    
    sns.barplot(data=df, x=x_col, y=y_col, hue=hue_col, 
                palette=custom_palette, edgecolor='black')
    
    plt.xlabel(x_col, fontsize=12)
    plt.ylabel(y_col, fontsize=12)
    plt.title(f"{y_col} vs {x_col} vs {hue_col}", fontsize=14)
    
    plt.legend(title=hue_col, bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()
```

#### **Hypothesis Testing**

**Hypothesis testing is a fundamental statistical technique used to make inferences about a population based on sample data. It helps determine whether an observed pattern or relationship in the data is due to chance or represents a real effect.**

```python
#Fuction to display styled dataframe
def format_dataframe(df):
    return df.style.set_properties(**{'text-align': 'center'}).set_table_styles(
        [{'selector': 'th', 'props': [('text-align', 'center')]}])
```

```python
#Categorical / discrete columns
categorical_cols = ['Gender', 'Department', 'Extracurricular_Activities', 'Age', 'Stress_Level (1-10)',
                    'Internet_Access_at_Home', 'Parent_Education_Level', 'Family_Income_Level']
```

```python
#Continuous columns
continuous_cols = ['Attendance (%)', 'Midterm_Score', 'Final_Score', 'Assignments_Avg', 'Quizzes_Avg', 
                  'Participation_Score', 'Projects_Score', 'Total_Score', 'Study_Hours_per_Week', 
                  'Sleep_Hours_per_Night']
```

<div style="text-align: left; padding: 20px;">
    <span style="color: #DF3E62; font-size: 23px; font-weight: bold;">Chi-square Test:
</span>
</div>

**The Chi-Square Test is a statistical method used to determine whether two categorical variables are related. It helps test the independence of variables by comparing observed and expected frequencies in a contingency table.**


#### **Chi-Square Value (χ²)**

**• The Chi-Square Value measures the difference between observed and expected counts**

**• A higher χ² value indicates a greater difference, suggesting a stronger relationship.**

 **• A lower χ² value means the observed data is close to the expected values, indicating independence.**


#### **P-Value**
 
**• The p-value helps determine statistical significance.**

 **• If p-value < 0.05 → The variables are likely dependent (reject the null hypothesis).**
 
 **• If p-value ≥ 0.05 → No significant relationship is found (fail to reject the null hypothesis).**

 

```python
#Chi-square test (categorical vs categorical)
def chi_square_test(col1, col2):
   contingency_table = pd.crosstab(df[col1], df[col2])
   chi2_stat, p, _, _ = stats.chi2_contingency(contingency_table)
   significance = 'Yes' if p < 0.05 else 'No'
   return pd.DataFrame([[col1, col2, chi2_stat, p, significance]], 
                        columns=['Feature 1', 'Feature 2', 'Chi-Square Value', 'P-Value', 'Significant Relationship'])

chi_square_results = pd.DataFrame(columns=['Feature 1', 'Feature 2', 'Chi-Square Value', 'P-Value', 'Significant Relationship'])

for i in range(len(categorical_cols)):
    for j in range(i+1, len(categorical_cols)):
        chi_square_results = pd.concat([chi_square_results, chi_square_test(categorical_cols[i], categorical_cols[j])], ignore_index=True)

print('Chi-Square Results')
display(format_dataframe(chi_square_results))
```

<div style="text-align: left; padding: 20px;">
    <span style="color: #DF3E62; font-size: 23px; font-weight: bold;">Spearman Correlation:
</span>
</div>

**The Spearman correlation is a non-parametric measure of the monotonic relationship between two continuous or ordinal variables. Unlike Pearson correlation, which assumes a linear relationship, Spearman assesses rank-based correlation, making it useful for non-linear relationships.**

#### **Interpreting Spearman Correlation (ρ)**
 **• ρ = +1 → Perfect positive monotonic relationship**
 
 **• ρ = -1 → Perfect negative monotonic relationship**
 
 **• ρ ≈ 0 → No monotonic relationship**


#### **P-Value**
 **• Determines statistical significance.**
 
 **• If p-value < 0.05 → The correlation is significant.**
 
 **• If p-value ≥ 0.05 → No significant correlation exists.**
 


```python
#Spearman correlation (continuous vs continuous)
def spearman_correlation(col1, col2):
    spearman_coeff, p = stats.spearmanr(df[col1], df[col2])
    significance = 'Yes' if p < 0.05 else 'No'
    return pd.DataFrame([[col1, col2, spearman_coeff, p, significance]], 
                        columns=['Feature 1', 'Feature 2', 'Spearman Coefficient', 'P-Value', 'Significant Relationship'])

spearman_results = pd.DataFrame(columns=['Feature 1', 'Feature 2', 'Spearman Coefficient', 'P-Value', 'Significant Relationship'])

for i in range(len(continuous_cols)):
    for j in range(i+1, len(continuous_cols)):
        spearman_results = pd.concat([spearman_results, spearman_correlation(continuous_cols[i], continuous_cols[j])], ignore_index=True)

print('Spearman Correlation Results')
display(format_dataframe(spearman_results))
```

<div style="text-align: left; padding: 20px;">
    <span style="color: #DF3E62; font-size: 23px; font-weight: bold;">ANOVA Test:
</span>
</div>

**ANOVA is a statistical test used to compare the means of three or more groups to determine if there is a statistically significant difference between them. It helps assess whether a categorical independent variable affects a continuous dependent variable.**

#### **F-Statistic**
 **• A higher F-value suggests that the difference between groups is more significant compared to random variation.**
 
 **• A lower F-value indicates that the group differences may be due to random chance.**

#### **P-Value**
 **• The p-value measures the probability that the observed differences are due to chance.**
 
 **• If p-value < 0.05 → At least one group is significantly different.**
 
 **• If p-value ≥ 0.05 → No significant difference between the groups.**


```python
#ANOVA test (continuous vs categorical)
def anova_test(col1, col2):
   groups = [df[col2][df[col1] == cat] for cat in df[col1].unique()]
   f_stat, p = stats.f_oneway(*groups)
   significance = 'Yes' if p < 0.05 else 'No'
   return pd.DataFrame([[col1, col2, f_stat, p, significance]], 
                        columns=['Feature 1', 'Feature 2', 'F-Statistic', 'P-Value', 'Significant Relationship'])

anova_results = pd.DataFrame(columns=['Feature 1', 'Feature 2', 'F-Statistic', 'P-Value', 'Significant Relationship'])

for cat_col in categorical_cols:
    for con_col in continuous_cols:
        anova_results = pd.concat([anova_results, anova_test(cat_col, con_col)], ignore_index=True)

print('ANOVA Results')
display(format_dataframe(anova_results))
```
